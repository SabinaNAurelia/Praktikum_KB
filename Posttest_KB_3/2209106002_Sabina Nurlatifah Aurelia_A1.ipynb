<h4>Visualisasi Data Posttest 3</h4>

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np

matplotlib.style.use('ggplot') # grafik menggunakan gaya ggplot
```

```python
df= pd.read_csv('tornados.csv')
df.head()
```

<h4>Menampilkan 5 data teratas</h4>

```python
#count the injuries in each year and plot the highest 10
injuries_sorted_by_year = df.groupby('yr')['inj'].sum().sort_values(ascending=False)
injuries_sorted_by_year[:42].plot(kind='bar', color=sns.color_palette('Paired'), figsize=(15, 7))
plt.xlabel('Tahun')
plt.ylabel('Cedera')
plt.title ('Banyaknya korban Cedera ')
plt.show()
```

<h4>Bencana Tornado di Amerika memiliki angka korban jiwa tertinggi di tahun 1974 dengan 5000+ jiwa</h4>

```python
sns.scatterplot(
    data=df,
    x='inj',
    y='fat',
    hue='yr',
)
plt.xlim(0, 50)
plt.ylim(0, 50)
```

<h4>Angka korban jiwa bencana tornado tertinggi ada pada tahun 1995, sisanya adalah korban cedera</h4>

```python
yr_group= df.groupby('yr').sum()
yr_group.head()
yr_group.index
```

<h4>membuat kolom year menjadi index</h4>

```python
sns.lineplot(
    data=yr_group.loc[:, 'mag':'fat'],
)
plt.xticks(np.arange(1950, 2020, 10).astype('str'))
```

<h4>Banyaknya orang yang cidera dan meninggal disebabkan oleh skala magnitudo dari tornado</h4>

```python
plt.pie(yr_group['inj'].sort_values(),
        labels= yr_group['inj'].sort_values().index,
        autopct= '%1.1f%%',
        startangle= 360)
```

<h4>Rata-rata presentase korban cidera tornado tertinggi ada pada tahun 1974 lebih besar dari tahun lainnya yaitu 7.0%</h4>

```python
year_injuries_corr = df.loc [:, 'stf':'fat'].corr()
year_injuries_corr
```

<h4>Menampilkan data antar korelasi dari <br> 
stf = state FIPS (Federal Information Processing Standards) number) <br>
mag = skala magnitudo<br>
inj = korban cidera <br>
fat = korban jiwa</h4>

```python
sns.heatmap(
    data=year_injuries_corr,
    annot=True
)
```
