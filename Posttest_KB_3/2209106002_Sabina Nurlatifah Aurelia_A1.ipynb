<h4>Visualisasi Data Posttest 3</h4>

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np

matplotlib.style.use('ggplot') # grafik menggunakan gaya ggplot
```

<h4>Import library yang dibutuhkan</h4>

```python
df= pd.read_csv('tornados.csv')
df.head()
```

<h4>Menampilkan 5 data teratas dari dataset</h4>

```python
#Total korban cidera berdasarkan tahun menggunakan diagram batang
injuries_sorted_by_year = df.groupby('yr')['inj'].sum().sort_values(ascending=True)
injuries_sorted_by_year[:73].plot(kind='bar', color=sns.color_palette('Paired'), figsize=(20, 10))
plt.xlabel('Year')
plt.ylabel('Injuried Person')
plt.title ('Banyaknya korban Cedera ')
plt.show()
```

<h4>Dari diagram diatas menunjukkan jumlah korban terluka berdasarkan tahunnya. 2018 merupakan angka terkecil dan angka yang terbesar ada pada tahun 1974</h4>

```python
# Histogram 
plt.hist(df['mag'], range=(1, 3))
plt.xlabel('Magnitude')
plt.ylabel('Total Injuried Person')
plt.title('Histogram of Magnitude')
plt.show()
```

<h4>Dari histogram diatas skala magnitudo jumlah orang yang terluka ada pada skala 1.00 dengan jumlah 2000+ orang terluka</h4>

```python
#scatterplot
sns.scatterplot(
    data=df,
    x='inj',
    y='fat',
    hue='yr',
)
plt.xlim(5, 100)
plt.ylim(5, 100)
```

<h4>Angka korban jiwa tertinggi ada pada sekitar tahun 1995 dimana hingga 20+ orang meninggal dunia, lalu tertinggi kedua sekitar tahun 2010 dengan jumlah korban jiwa sekitar 20+ orang</h4>

```python
yr_group= df.groupby('yr').sum()
yr_group.head()
yr_group.index
```

<h4>Membuat kolom year menjadi index</h4>

```python
#lineplot
sns.lineplot(data=yr_group.loc[:, 'mag':'fat'])
plt.xticks(np.arange(1950, 2020, 10).astype('str'))
```

<h4>Angka kematian mulai tinggi sekitar tahun 1950 dengan skala magnitudo yang tidak terlalu tinggi,  <br>tetapi banyak korban yang cidera</h4>

```python
#pie chart
plt.pie(yr_group['inj'].sort_values(),
        labels= yr_group['inj'].sort_values().index,
        autopct= '%1.1f%%',
        startangle= 360)
```

<h4>Rata-rata persentase korban cidera bencana tornado tertinggi ada pada tahun 1974 dengan total 7.0% dari jumlah penduduk America/Chicago</h4>

```python
year_injuries_corr = df.loc [:, 'stf':'fat'].corr()
year_injuries_corr
```

<h4>Menampilkan data antar korelasi dari <br> 
stf = state FIPS (Federal Information Processing Standards) <br>
mag = skala magnitudo<br>
inj = korban cidera <br>
fat = korban jiwa</h4>

```python
#heatmap
sns.heatmap(data=year_injuries_corr, cmap='Reds', annot=True, center=0)
plt.show()
```

<h4>stf = state FIPS (Federal Information Processing Standarts) <br>
mag = Skala Magnitudo<br>
inj = Korban Cidera <br>
fat = Korban Jiwa <br>
Korelasi stf lebih rendah dibandingkan dengan jumlah korban jiwa yang terjadi selama bencana tornado 1950-2022</h4>
