<h5>Menampilkan 10 record pertama dari dataset dengan attribute bertipe angka</h5>

```python
dataph = {}

dataph['Jumlah Record'] = ph.shape[0]
dataph['Jumlah Attribute'] = ph.shape[1]
dataph['Nama Attribute'] = ph.columns.tolist()
dataph['Jumlah Attribute Dengan Tipe Angka'] = ph.select_dtypes(['int64', 'float64']).shape[1]

dictionarykey = list(dataph.keys())

table= prtl.PrettyTable()
table.add_column(dictionarykey[0], [dataph['Jumlah Record']])
table.add_column(dictionarykey[1], [dataph['Jumlah Attribute']])
table.add_column(dictionarykey[2], ['\n'.join(dataph['Nama Attribute'])])
table.add_column(dictionarykey[3], [dataph['Jumlah Attribute Dengan Tipe Angka']])
print(table)
```
